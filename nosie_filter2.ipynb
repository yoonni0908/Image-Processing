```python
src = cv2.imread('Homework #3 images\High-boost filer images\Fig0327(a)(tungsten_original)_hb.jpg')
#Convert to grayscale
img_gray = cv2.cvtColor(src, cv2.COLOR_BGR2GRAY)
#Blur the image for better edge detection
img_blur = cv2.GaussianBlur(img_gray, (3,3), 0)

#sobel
#combined x and y sobel edge detection
sobel_3 = cv2.Sobel(img_blur , ddepth = -1, dx = 1, dy = 1, delta = 128, ksize = 3)
sobel_5 = cv2.Sobel(img_blur , ddepth = -1, dx = 1, dy = 1, delta = 128, ksize = 5)
sobel_7 = cv2.Sobel(img_blur , ddepth = -1, dx = 1, dy = 1, delta = 128, ksize = 7)

cv2.imshow('src', cv2.resize(src, dsize=(0, 0), fx=0.5, fy=0.5, interpolation=cv2.INTER_LINEAR))
cv2.imshow('dst_sobel_3', cv2.resize(sobel_3, dsize=(0, 0), fx=0.5, fy=0.5, interpolation=cv2.INTER_LINEAR))
cv2.imshow('dst_sobel_5', cv2.resize(sobel_5, dsize=(0, 0), fx=0.5, fy=0.5, interpolation=cv2.INTER_LINEAR))
cv2.imshow('dst_sobel_7', cv2.resize(sobel_7, dsize=(0, 0), fx=0.5, fy=0.5, interpolation=cv2.INTER_LINEAR))
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
src = cv2.imread('Homework #3 images\Color noisy image\lena_noise.png')

img_gray = cv2.cvtColor(src, cv2.COLOR_BGR2GRAY)
img_blur = cv2.GaussianBlur(img_gray, (3,3), 0)

#prewitt
kernelx = np.array([[1,1,1],[0,0,0],[-1,-1,-1]])
kernely = np.array([[-1,0,1],[-1,0,1],[-1,0,1]])
prewittx = cv2.filter2D(img_blur, -1, kernelx)
prewitty = cv2.filter2D(img_blur, -1, kernely)
prewitt = prewittx + prewitty

cv2.imshow('src', cv2.resize(src, dsize=(0, 0), fx=0.5, fy=0.5, interpolation=cv2.INTER_LINEAR))
cv2.imshow('dst_Prewitt_X', cv2.resize(prewittx, dsize=(0, 0), fx=0.5, fy=0.5, interpolation=cv2.INTER_LINEAR))
cv2.imshow('dst_Prewitt_Y', cv2.resize(prewitty, dsize=(0, 0), fx=0.5, fy=0.5, interpolation=cv2.INTER_LINEAR))
cv2.imshow('dst_Prewitt', cv2.resize(prewitt, dsize=(0, 0), fx=0.5, fy=0.5, interpolation=cv2.INTER_LINEAR))
           
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
src = cv2.imread('Homework #3 images/Color noisy image/lena_noise.png')

img_gray = cv2.cvtColor(src, cv2.COLOR_BGR2GRAY)
img_blur = cv2.GaussianBlur(img_gray, (3,3), 0)

#canny edge detection
canny = cv2.Canny(img_blur, threshold1 = 50, threshold2 = 100)

cv2.imshow('src', cv2.resize(src, dsize=(0, 0), fx=0.5, fy=0.5, interpolation=cv2.INTER_LINEAR))
cv2.imshow('dst_canny', cv2.resize(canny, dsize=(0, 0), fx=0.5, fy=0.5, interpolation=cv2.INTER_LINEAR))
cv2.waitKey(0)
cv2.destroyAllWindows()
```
